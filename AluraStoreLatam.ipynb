### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda1 = pd.read_csv ( url )
tienda2 = pd.read_csv ( url2 )
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

#tienda.head()

print(" Datos cargados con éxito.")
```

#1. Análisis de facturación



```python
# Sumar la columna "Precio" para saber cuánto vendió cada tienda
facturacion1 = tienda1["Precio"].sum()
facturacion2 = tienda2["Precio"].sum()
facturacion3 = tienda3["Precio"].sum()
facturacion4 = tienda4["Precio"].sum()

# Mostrar los resultados
print("Facturación tienda 1:", facturacion1)
print("Facturación tienda 2:", facturacion2)
print("Facturación tienda 3:", facturacion3)
print("Facturación tienda 4:", facturacion4)
```

```python
import matplotlib.pyplot as plt  # Librería para gráficos

# Nombres de las tiendas
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

# Ventas totales por tienda (asegúrate de tener estas variables definidas)
ventas = [facturacion1, facturacion2, facturacion3, facturacion4]

# Crear la gráfica
plt.figure(figsize=(8, 5))  # Tamaño del gráfico
plt.bar(tiendas, ventas, color='skyblue')  # Gráfico de barras
plt.title('Facturación total por tienda')  # Título del gráfico
plt.ylabel('Valor vendido (en pesos)')     # Etiqueta del eje Y
plt.xlabel('Tiendas')                      # Etiqueta del eje X
plt.grid(axis='y', linestyle='--', alpha=0.5)  # Líneas de guía horizontales
plt.show()  # Mostrar el gráfico
```

# 2. Ventas por categoría

```python
# Ventas por categoría
'''
En este debes calcular la cantidad de productos vendidos por categoría en cada tienda.
La idea es agrupar los datos por categoría y contar el número de ventas de cada tipo,
mostrando las categorías más populares de cada tienda.
'''

ventas_por_categoria1 = tienda1["Categoría del Producto"].value_counts().reset_index() #.reset_index() elimina Name: count, dtype: int64
ventas_por_categoria1.columns = ['Categoría', 'Cantidad']

ventas_por_categoria2 = tienda2["Categoría del Producto"].value_counts().reset_index()
ventas_por_categoria2.columns = ['Categoría', 'Cantidad']

ventas_por_categoria3 = tienda3["Categoría del Producto"].value_counts().reset_index()
ventas_por_categoria3.columns = ['Categoría', 'Cantidad']

ventas_por_categoria4 = tienda4["Categoría del Producto"].value_counts().reset_index()
ventas_por_categoria4.columns = ['Categoría', 'Cantidad']


print("Ventas Tienda 1:\n", ventas_por_categoria1, "\n")
print("Ventas Tienda 2:\n", ventas_por_categoria2, "\n")
print("Ventas Tienda 3:\n", ventas_por_categoria3, "\n")
print("Ventas Tienda 4:\n", ventas_por_categoria4, "\n")
```

# 3. Calificación promedio de la tienda


```python
promedio_t1 = tienda1["Calificación"].mean()
promedio_t2 = tienda2["Calificación"].mean()
promedio_t3 = tienda3["Calificación"].mean()
promedio_t4 = tienda4["Calificación"].mean()

# Mostrar resultados
print(" Promedio de calificación - Tienda 1:", promedio_t1)
print(" Promedio de calificación - Tienda 2:", promedio_t2)
print(" Promedio de calificación - Tienda 3:", promedio_t3)
print(" Promedio de calificación - Tienda 4:", promedio_t4)
```

```python
# Librería para gráficos
import matplotlib.pyplot as plt

# Datos
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
calificaciones = [promedio_t1, promedio_t2, promedio_t3, promedio_t4]

# Crear gráfico de barras
plt.figure(figsize=(8, 5))
plt.bar(tiendas, calificaciones, color='gold')
plt.title('Promedio de Calificación por Tienda')
plt.ylabel('Calificación promedio (de 1 a 5)')
plt.ylim(0, 5)  # Limitar el eje Y para ver bien la escala de 1 a 5
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()
```

# 4. Productos más y menos vendidos

```python
# Tienda 1
productos_t1 = tienda1["Producto"].value_counts()
mas_vendido_t1 = productos_t1.idxmax()
menos_vendido_t1 = productos_t1.idxmin()

# Tienda 2
productos_t2 = tienda2["Producto"].value_counts()
mas_vendido_t2 = productos_t2.idxmax()
menos_vendido_t2 = productos_t2.idxmin()

# Tienda 3
productos_t3 = tienda3["Producto"].value_counts()
mas_vendido_t3 = productos_t3.idxmax()
menos_vendido_t3 = productos_t3.idxmin()

# Tienda 4
productos_t4 = tienda4["Producto"].value_counts()
mas_vendido_t4 = productos_t4.idxmax()
menos_vendido_t4 = productos_t4.idxmin()

# Mostrar los resultados
print(" Productos más y menos vendidos por tienda:\n")
print(f"Tienda 1 ➜ Más vendido: {mas_vendido_t1} | Menos vendido: {menos_vendido_t1}")
print(f"Tienda 2 ➜ Más vendido: {mas_vendido_t2} | Menos vendido: {menos_vendido_t2}")
print(f"Tienda 3 ➜ Más vendido: {mas_vendido_t3} | Menos vendido: {menos_vendido_t3}")
print(f"Tienda 4 ➜ Más vendido: {mas_vendido_t4} | Menos vendido: {menos_vendido_t4}")

```

# 5. Envío promedio por tienda

```python

# Valor del envío promedio por tienda
'''
En este paso, debes calcular el costo de envío promedio para cada tienda.
El objetivo es comprender cuánto se gasta, en promedio, en el envío de cada tienda.
'''

envio_promedio1 = tienda1["Costo de envío"].mean()
envio_promedio2 = tienda2["Costo de envío"].mean()
envio_promedio3 = tienda3["Costo de envío"].mean()
envio_promedio4 = tienda4["Costo de envío"].mean()

print(f"El envío promedio de la tienda 1 es de: {envio_promedio1:.2f}")
print(f"El envío promedio de la tienda 2 es de: {envio_promedio2:.2f}")
print(f"El envío promedio de la tienda 3 es de: {envio_promedio3:.2f}")
print(f"El envío promedio de la tienda 4 es de: {envio_promedio4:.2f}")
```

```python
# Gráfico de torta: Valor del envío promedio por tienda

tiendas = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]
envios = [envio_promedio1, envio_promedio2, envio_promedio3, envio_promedio4]

plt.figure(figsize=(7, 7))
plt.pie(envios, labels=tiendas, autopct='%1.1f%%', startangle=90, colors=plt.cm.Set3.colors)
plt.title("Proporción del Costo Promedio de Envío por Tienda")
plt.axis('equal')  # Hace que el gráfico sea circular
plt.tight_layout()
plt.show()
```
